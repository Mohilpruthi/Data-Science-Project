### Data checking and cleaning

Firstly lets check for duplicate values

```python
data[data.duplicated()]
```

No duplicates

Let's look if we have any null values

```python
data.isnull().sum()
```

Now, we will remove Anime with no rating as mostly unrealeased titles are the ones with no rating
and it doesn't seem reasonable for us to work with unreleased anime

```python
data = data[data.Rating.notna()]
data.isnull().sum()
```

```python
data.shape
```

Next, we will remove the Anime having Specials as their type, because specials are only an addition to the original work and are unlikely to be of a value without the original. Just in case, we will include them in a separate frame.

```python
data_special = data[data.Type=='Special']
data=data[data.Type!='Special']
data
```

<br>

Next, we will update the data for anime with an unknown number of episodes (unreleased, or planned for release). For titles like One Piece (which are still releasing), we'll just use the number of episodes available at the moment.

```python
data[data.Episodes=='Unknown']
```

We will update these unknown values with the latest no. of episodes available onilne

```python
data.Episodes[27]=1004
data.Episodes[197]=231
data.Episodes[502]=12
data.Episodes[592]=1032
data.Episodes[686]=12
data.Episodes[891]=12
data.Episodes[1347]=12
data[data.Episodes=='Unknown']
```

<br>

Next we will convert the 'Followers' and 'Episodes' columns to a numeric format

```python
data.Episodes=pd.to_numeric(data.Episodes)
data.Followers = data.Followers.replace(',', '', regex=True)
data.Followers=pd.to_numeric(data.Followers)
data.info()
```

<br>

Finally, now our dataset looks like this, and the number of records has been reduced from 2000 to 1889.

```python
data
```

## Distribution by rating

Now, let's break the anime into categories and see their ratings.

```python
data['Rating'].plot(kind='hist', title='Distribution of anime by rating', color="orange")
plt.xlabel('Ratings')
plt.ylabel("No. of Anime")
plt.show()    
```

This distribution seems normal Gaussian (bell curve)

<br>

## Correlation between rating and genres and followers and genres

```python
data2 = data.assign(Genre=data['Genre'].str.split(',')).explode('Genre') 
data2
```

```python
data2_rating_r = data2.groupby('Genre').describe()['Rating'].sort_values('mean', ascending=False)
data2_rating_r
```

```python
data2_rating_r.plot(kind='barh', y='mean', figsize=(10,7), color="green", title='Average anime rating by genres').invert_yaxis()
```

Observing Anime ratings by genres, we found that suspense, sports and drama are most popular genres but there is not much diffrence of rating in Top 10 genres as we would think.

<br>

```python
data2_rating_f = data2.groupby('Genre').describe()['Followers'].sort_values('mean', ascending=False)
data2_rating_f
```

<br>

```python
data2_rating_f.plot(kind='barh', y='mean', figsize=(10,7), color="m",title='Average number of followers but different genres').invert_yaxis()
```

Moreover 'Suspense' is clearly in the lead in no. of followers by a wide margin.

<br>

## Distribution by audience age

```python
data.Age.value_counts().plot(kind='bar', figsize=(10,5), color = "c",title='Anime distribution by audience age');
```

Observing age distribution of Anime, we can see that more than half of the entire industry works for teenagers

<br>

## Distribution by ages of auditory and number of followers

```python
data_ages = data.groupby('Age').describe()['Followers'].sort_values('count', ascending=False)[:5].sort_values('mean')
data_ages
```

```python
data_ages.plot(kind='barh', y='mean', figsize=(8,5), color="y",title='Distribution of anime by age categories and number of followers');
```

Obseving age as well as followers distrubution together, it's interesting to note that although the majority of industry is working for teenagers Children's anime remains below the highest age group, while more adult ones stay higher (one explanation could be that even if a child watches anime as a child' age, it is unlikely that they will follow it, while as age increases, this probability will become higher).

<img src="levi.jpg" align="right" width="500" height="450">

```python
data_ages_2 = data.groupby('Age').describe()['Episodes'].sort_values('count', ascending=False)[:5].sort_values('mean')
data_ages_2
```

```python
data_ages_2.plot(kind='barh', y='mean', color="red", title='Distribution of anime by age categories and number of episodes');
```

Another interesting observation can be made by looking at the average number of episodes - for a children's audience it's much higher (one explanation could be that due to the fact that children are less demanding on the content they receive, it's much easier to show them more episodes of the same series).

<br>

## Distributon by types of anime

```python
data.Type.value_counts()[:5].plot(kind='pie', figsize=(10,5), title='Distribution of anime by type');
```

Observing distribution by type of anime, there is no surprise that TV series it the most common type of anime.

<br>

## Distribution by episodes

```python
bins_episodes = [0,1,11,14,30,60,1050]
data_episodes = pd.cut(data.Episodes, bins_episodes).value_counts()[:].sort_index(ascending=False)
data_episodes
```

```python
data_episodes[:10].plot(kind='bar', figsize=(10,5), color="hotpink", title = 'Anime distribution by episodes');
```

Observing distribution by no. of episodes, the standard 12-episode format has a solid lead. At the same time, the number of studies wishing to release more than 30 episodes is lower.

<br>

## Distribution by studios

```python
data_studios = data.Study.str.get_dummies(sep=',')
data_studios = data_studios.sum().reset_index().sort_values(0, ascending=False).set_index('index')
data_studios[:20]
```

```python
data_studios[:20].plot(kind='barh', figsize=(10,5), title='Distribution of anime by animation studios', legend=False).invert_yaxis()
```

Observing distribution by studios, top 3 leading studios are J.C Staff, Madhouse and A-1 Pictures
